##### Коэффициент детерминации, также называемый оценкой R2, используется для оценки производительности модели линейной регрессии. Это величина изменения в выходном зависимом атрибуте, которая предсказуема из входных независимых переменных. Он используется для проверки того, насколько хорошо наблюдаемые результаты воспроизводятся моделью, в зависимости от отношения общего отклонения результатов, описанных моделью.

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test["price"], y_pred.flatten())
```
